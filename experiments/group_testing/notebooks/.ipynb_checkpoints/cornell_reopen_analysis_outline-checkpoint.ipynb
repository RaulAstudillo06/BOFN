```python
def plot_cip_thresholds(df_cips, threshold=0.1, xlabel="", title="", figsize=(10,6)):
    xs = sorted(list(df_cips.keys()))
    p_thresholds = []
    for x in xs:
        cips = extract_cips(df_cips[x])
        cip_exceed_thresh = [cip for cip in cips if cip >= threshold]
        p_thresholds.append(len(cip_exceed_thresh) / len(cips))
    plt.figure(figsize=figsize)
    plt.plot(xs, p_thresholds, marker='o')
    plt.xlabel(xlabel)
    plt.ylabel("Probability CIP Exceeds {}%".format(threshold * 100))
    plt.title(title)
    plt.show()

plot_cip_thresholds(dfs_ctc_varied, xlabel="Contact Tracing Constant")
```
