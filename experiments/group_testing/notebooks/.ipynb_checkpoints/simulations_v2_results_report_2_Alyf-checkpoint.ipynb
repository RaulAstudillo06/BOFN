```python
def sim_many_test_days(base_params, test_days, use_SIR_dynamics=False, time_horizon=250, only_bucket_test=True):
    dfs_cts = {}
    dfs_bucket = {}
    base_params_cts = base_params.copy()
    base_params_cts['days_between_tests'] = 1
    
    base_params_bucket = base_params.copy()
    base_params_bucket['test_population_fraction'] = 1
    
    for T in test_days:
        if not only_bucket_test:
            base_params_cts['test_population_fraction'] = 1/T
            if use_SIR_dynamics:
                sim_cts = SIRStochasticSimulation(base_params_cts)
            else:
                sim_cts = StochasticSimulation(base_params_cts)
            dfs_cts[T] = []
            for _ in range(100):
                dfs_cts[T].append(sim_cts.run_new_trajectory(time_horizon))
        
        base_params_bucket['days_between_tests'] = T
        if use_SIR_dynamics:
            sim_bucket = SIRStochasticSimulation(base_params_bucket)
        else:
            sim_bucket = StochasticSimulation(base_params_bucket)
        dfs_bucket[T] = []
        for _ in range(100):
            dfs_bucket[T].append(sim_bucket.run_new_trajectory(time_horizon))
        
        print("Done test frequency {}".format(T))
    
    return dfs_cts, dfs_bucket
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.size'] = 12

def add_plot(df, params, color='blue', ID_only=False):
    if ID_only:
        cols = ['ID_{}'.format(x) for x in range(params['max_time_ID'])]
    else:
        cols = ['ID_{}'.format(x) for x in range(params['max_time_ID'])] + \
                ['pre_ID_{}'.format(x) for x in range(params['max_time_pre_ID'])]
    plt.plot(df[cols].sum(axis=1), linewidth=10.0, alpha=0.1, color=color)    


# just need params to get the proper column names
def plot_sims(dfs1, dfs2, title1, title2, params, ID_only=False):
    plt.figure(figsize=(20,12))

    plt.subplot(2,1,1)
    
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title1)

    for df in dfs1:
        add_plot(df, params, ID_only=ID_only)

    plt.subplot(2,1,2)
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title2)
    for df in dfs2:
        add_plot(df, params, color='green', ID_only=ID_only)

    plt.show()

def plot_many_sims(dfs_cts, dfs_bucket, params, ID_only):
    for test_frequency in dfs_cts.keys():
        df_cts = dfs_cts[test_frequency]
        df_bucket = dfs_bucket[test_frequency]
        title_cts = "Trajectory Under Continuous Testing: Test 1/{} Population Every Day".format(test_frequency)
        title_bucket = "Trajectory Under Bucket Testing: Test Everyone Every {} Days".format(test_frequency)
        plot_sims(df_cts, df_bucket, title_cts, title_bucket, params, ID_only)
```

```python
import numpy as np
from scipy.stats import geom, poisson

import functools

@functools.lru_cache(maxsize=128)
def poisson_pmf(max_time, mean_time):
    pmf = list()
    for i in range(1, max_time):
        pmf.append(poisson.pmf(i, mean_time))
    pmf.append(1-np.sum(pmf))
    return np.array(pmf)

```

```python
plt.rcParams['font.size'] = 12

def add_plot(df, params, color='blue', ID_only=False):
    if ID_only:
        cols = ['ID_{}'.format(x) for x in range(params['max_time_ID'])]
    else:
        cols = ['ID_{}'.format(x) for x in range(params['max_time_ID'])] + \
                ['pre_ID_{}'.format(x) for x in range(params['max_time_pre_ID'])]
    plt.plot(df[cols].sum(axis=1), linewidth=10.0, alpha=0.1, color=color)    


# just need params to get the proper column names
def plot_all_config_sims(dfs1, dfs2, dfs3, dfs4, dfs5, dfs6, title1, title2, title3, title4, title5, title6, params, ID_only=False):
    plt.figure(figsize=(20,36))

    plt.subplot(6,1,1)
    
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title1)

    for df in dfs1:
        add_plot(df, params, ID_only=ID_only)

    plt.subplot(6,1,2)
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title2)
    for df in dfs2:
        add_plot(df, params, color='green', ID_only=ID_only)

    plt.subplot(6,1,3)
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title3)
    for df in dfs3:
        add_plot(df, params, color='green', ID_only=ID_only)

    plt.subplot(6,1,4)
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title4)
    for df in dfs4:
        add_plot(df, params, color='green', ID_only=ID_only)

    plt.subplot(6,1,5)
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title5)
    for df in dfs5:
        add_plot(df, params, color='green', ID_only=ID_only)

    plt.subplot(6,1,6)
    plt.xlabel("Day")
    plt.ylabel("Number of Active Infections")
    plt.title(title6)
    for df in dfs6:
        add_plot(df, params, color='green', ID_only=ID_only)
        
    plt.show()

def plot_all_configurations(dfs_no_intervention, dfs_ct_only, dfs_testing_cts, dfs_testing_bucket, dfs_everything_cts, dfs_everything_bucket, params, ID_only):
    for test_frequency in dfs_testing_cts.keys():
        df_no_intervention = dfs_no_intervention[0]
        df_ct_only = dfs_ct_only[0]
        df_testing_cts = dfs_testing_cts[test_frequency]
        df_testing_bucket = dfs_testing_bucket[test_frequency]
        df_everything_cts = dfs_everything_cts[test_frequency]
        df_everything_bucket = dfs_everything_bucket[test_frequency]
        
        title_no_intervention = "Trajectory Under No Interventions"
        title_ct_only = "Trajectory Under Contact Tracing Only"
        title_testing_cts = "Trajectory Under Continuous Testing: Test 1/{} Population Every Day".format(test_frequency)
        title_testing_bucket = "Trajectory Under Bucket Testing: Test Everyone Every {} Days".format(test_frequency)
        title_everything_cts = "Trajectory Under Contact Tracing and Continuous Testing: Test 1/{} Population Every Day".format(test_frequency)
        title_everything_bucket = "Trajectory Under Contact Tracing and Bucket Testing: Test Everyone Every {} Days".format(test_frequency)
        
        plot_all_config_sims(df_no_intervention, df_ct_only, df_testing_cts, df_testing_bucket, df_everything_cts, df_everything_bucket, title_no_intervention, title_ct_only, title_testing_cts, title_testing_bucket, title_everything_cts, title_everything_bucket, params, ID_only)
```

```python
def sim_all_configs(base_params, test_days, use_SIR_dynamics=False, time_horizon=250, only_bucket_test=False):
    dfs_no_intervention = {}
    dfs_ct_only = {}
    dfs_testing_cts = {}
    dfs_testing_bucket = {}
    dfs_everything_cts = {}
    dfs_everything_bucket = {}

    params_no_intervention = base_params.copy()
    params_no_intervention['perform_contact_tracing'] = False
    params_no_intervention['days_between_tests'] = time_horizon+100
    
    params_ct_only = base_params.copy()
    params_ct_only['perform_contact_tracing'] = True
    params_ct_only['days_between_tests'] = time_horizon+100
    
    params_testing_cts = base_params.copy()
    params_testing_cts['perform_contact_tracing'] = False
    params_testing_cts['days_between_tests'] = 1
    
    params_testing_bucket = base_params.copy()
    params_testing_bucket['perform_contact_tracing'] = False
    params_testing_bucket['test_population_fraction'] = 1
    
    params_everything_cts = base_params.copy()
    params_everything_cts['perform_contact_tracing'] = True
    params_everything_cts['days_between_tests'] = 1
    
    params_everything_bucket = base_params.copy()
    params_everything_bucket['perform_contact_tracing'] = False
    params_everything_bucket['test_population_fraction'] = 1
    
#     base_params_cts['days_between_tests'] = 1
    
#     base_params_bucket = base_params.copy()
#     base_params_bucket['test_population_fraction'] = 1
    
    for T in test_days:
        if not only_bucket_test:
            params_testing_cts['test_population_fraction'] = 1/T
            params_everything_cts['test_population_fraction'] = 1/T
            if use_SIR_dynamics:
                sim_testing_cts = SIRStochasticSimulation(params_testing_cts)
                sim_everything_cts = SIRStochasticSimulation(params_everything_cts)
            else:
                sim_testing_cts = StochasticSimulation(params_testing_cts)
                sim_everything_cts = StochasticSimulation(params_everything_cts)
            dfs_testing_cts[T] = []
            dfs_everything_cts[T] = []
            for _ in range(100):
                dfs_testing_cts[T].append(sim_testing_cts.run_new_trajectory(time_horizon))
                dfs_everything_cts[T].append(sim_everything_cts.run_new_trajectory(time_horizon))
        
        params_testing_bucket['days_between_tests'] = T
        params_everything_bucket['days_between_tests'] = T
        if use_SIR_dynamics:
            sim_testing_bucket = SIRStochasticSimulation(params_testing_bucket)
            sim_everything_bucket = SIRStochasticSimulation(params_everything_bucket)
        else:
            sim_testing_bucket = StochasticSimulation(params_testing_bucket)
            sim_everything_bucket = StochasticSimulation(params_everything_bucket)
        dfs_testing_bucket[T] = []
        dfs_everything_bucket[T] = []
        for _ in range(100):
            dfs_testing_bucket[T].append(sim_testing_bucket.run_new_trajectory(time_horizon))
            dfs_everything_bucket[T].append(sim_everything_bucket.run_new_trajectory(time_horizon))
        
        print("Done test frequency {}".format(T))
        
    params_no_intervention
    params_ct_only
    if use_SIR_dynamics:
        sim_no_intervention = SIRStochasticSimulation(params_no_intervention)
        sim_ct_only = SIRStochasticSimulation(params_ct_only)
    else:
        sim_no_intervention = StochasticSimulation(params_no_intervention)
        sim_ct_only = StochasticSimulation(params_ct_only)
    dfs_no_intervention[0] = []
    dfs_ct_only[0] = []
    for _ in range(100):
        dfs_no_intervention[0].append(sim_no_intervention.run_new_trajectory(time_horizon))
        dfs_ct_only[0].append(sim_ct_only.run_new_trajectory(time_horizon))
    
    return dfs_no_intervention, dfs_ct_only, dfs_testing_cts, dfs_testing_bucket, dfs_everything_cts, dfs_everything_bucket
```
