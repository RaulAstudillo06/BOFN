```python
sql_query = """
select *
    from county_data c
        inner join (
            select UID - 84000000 as fips,
                    date,
                    admin2 as county_name,
                    combined_key as long_county_name,
                    province_state,
                    count as death_count,
                    population
            from time_series_covid19_deaths
        ) t
    on c.fips = t.fips;
"""
cur = conn.cursor()
cur.execute(sql_query)
rows = cur.fetchall()
```

```python
sql_query = """
select date,
        county_name,
        long_county_name,
        province_state as state,
        POP060210 as pop_density,
        population,
        death_count
    from county_data c
        inner join (
            select UID - 84000000 as fips,
                    date,
                    admin2 as county_name,
                    combined_key as long_county_name,
                    province_state,
                    count as death_count,
                    population
            from time_series_covid19_deaths
        ) t
    on c.fips = t.fips;
"""
cur = conn.cursor()
cur.execute(sql_query)
rows = cur.fetchall()
```
